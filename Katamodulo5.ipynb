## Ejercicio 2: Convertir cadenas en numeros y usar valores absolutos

```python
# Almacenar las entradas del usuario
#Pista: variable = input("¿Cuál es tu nombre?")

Planeta1 = input("Distancia desde el Sol al primer planeta (km)")
Planeta2 = input("Distancia desde el Sol al segundo planeta (km)")

planetaA = input("Nombre del primer planeta")
planetaB = input("Nombre del segundo planeta")

Planeta1 = int(Planeta1)
Planeta2 = int(Planeta2)

Distancia_dif = abs(Planeta1 - Planeta2) 
Distancia_milla = int(Distancia_dif * 0.621)

print(f"Hay una distancia de {Distancia_dif} km entre {planetaA.upper()} y {planetaB.upper()}")
print(f"Hay una distancia de {Distancia_milla} millas entre {planetaA.upper()} y {planetaB.upper()}")
```
