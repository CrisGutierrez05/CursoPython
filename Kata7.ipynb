```python
#Variable que almacena el texto
user_input = ''
#Lista que almacena los textos del usuario
inputs = []
#Ciclo while
while user_input.lower() != 'done':
    #Verificar si hay valor en user_input
    if user_input:
        #Almacenar el valor en la lista
        inputs.append(user_input)
    #Capturar un nuevo valor
    user_input = input('Enter a new value, or done when done')
print(inputs)
```

## Ejercicio 1: CreaciÃ³n de un bucle "while"

```python
# Uso de ciclos while en Python
#Declara dos variables
#Ciclo while, donde se ejecute hasta que aparezca done en new_planet, comprobar si la variable contiene un valor que sea el nombre de un planeta, en caso de que si agregar el valor a la variable planets. Usar input para solicitar qu eingrese un nuevo nombre de planeta o done si termino, y almacenar el valor de input en la nueva variable.

new_planet = ''

planets = []

while new_planet.lower() != 'done':
    if new_planet:
        planets.append(new_planet)
    new_planet = input('Introduzca el nombre de un planeta (valor nuevo), o escriba done si ha terminado')
print(planets)
```

## Ejercicio 2: CreaciÃ³n de un ciclo "for"

```python
#Ciclo para una lista.
#Crear un ciclo que muestre las entradas del usuario.
new_planet = ''

planets = []

while new_planet.lower() != 'done':
    if new_planet:
        planets.append(new_planet)
    new_planet = input('Introduzca el nombre de un planeta (valor nuevo), o escriba done si ha terminado')

from time import sleep
for new_planet in planets:
    print(new_planet)
    sleep(1)
```
