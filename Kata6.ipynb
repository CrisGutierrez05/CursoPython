## Ejercicio 2: Trabajando con datos de una lista

```python
# Lista de Planetas
planetas = ['Mercurio', 'Venus', 'Tierra', 'Marte', 'Jupiter', 'Saturno', 'Urano', 'Neptuno']
print('Hay', len(planetas), 'planetas')

nombre_planeta = input('Introduzca el Nombre del planeta empezando con mayúscula')
planeta = planetas.index(nombre_planeta)

planetas_antes_planeta = planetas[0:planeta]
print(f'El planeta {nombre_planeta} se encuentra en la posición', planeta + 1,' con respecto al Sol, también estos planetas son más cercanos al Sol', planetas_antes_planeta)

# Muestra los planetas más lejanos al sol
planetas_despues_planeta = planetas[planeta + 1:]
print('Y estos planetas son más lejanos al Sol', planetas_despues_planeta)
```
