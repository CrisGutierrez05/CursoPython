```python
def sequence_time(*args):
    total_minutes = sum(args)
    if total_minutes < 60:
        return f'Total time to launch is {total_minutes} minutes'
    else: 
        return f'Total time to launch is {total_minutes/60} hours'
sequence_time(4, 14, 48)        
```

```python
# Función con un informe preciso de la misión. Considera hora de prelanzamiento, tiempo de vuelo, destino, tanque externo y tanque interno

def reporte_de_misión(*hora_prelanzamiento, tiempo_vuelo, destino, tanque_ext, tanque_int):
    return f"""
    Misión a {destino}
    Tiempo total: {hora_prelanzamiento + tiempo_vuelo} minutos
    Combustible: {tanque_int + tanque_ext} litros
    """
print(reporte_de_misión(3, 40, "Marte", 4000, 20000))

```

```python
def reporte_de_misión(destinación, *tiempo, **combustible):
    return f"""
    Misión con destino a {destinación}
    Tiempo total de misión: {sum(tiempo)} minutos
    Combustible disponible: {sum(combustible.values())} litros
    """
print(reporte_de_misión("Luna", 15, 30, 70, interno=100000000, externo=80000000))
```

```python
def reporte_de_misión(destinación, *tiempo, **combustible):
    reporte_final = f"""
    Misión hacia {destinación} {('-'*90)}
    Tiempo total de viaje: {sum(tiempo)} minutos
    Combustible total: {sum(combustible.values())}
    """
    for tanque_interno, litros in combustible.items():
        reporte_final += f"{tanque_interno} tanque --> {litros} litros disponibles\n"
    return reporte_final

print(reporte_de_misión("Luna", 14, 20, 35, tanque_interno=14500000, tanque_externo=12000000))
```
